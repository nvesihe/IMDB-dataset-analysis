## IMDB top 1000 movies dataset analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imdb_top_1000.csv')

pd.set_option('display.max.columns', 16)

df.head(10)
```

```python
#Remove unnecessary columns.
df2 = df.drop(columns=['Poster_Link', 'Certificate', 'Meta_score', 'No_of_Votes', 'Overview', 'Gross' ])

df2.rename(columns={"Released_Year": "Release_Year"}, inplace=True)

df2.head()
```

```python
# Check info about the data
df2.info()
```

```python
# Check for Null-values
df2.isnull().sum()

```

### Top 20 genres in IMDB top movies

```python
# First split genres into individual rows


# Count occurrences of each genre
genre_counts = df2['Genre'].value_counts()

top_20_genres = genre_counts.head(20)

import seaborn as sns

plt.rcParams['figure.dpi'] = 200

top_20_genres_df = top_20_genres.reset_index()
top_20_genres_df.columns = ['Genre', 'Count']

sns.barplot(data=top_20_genres_df, y='Genre', x='Count', palette='viridis')
plt.title('Top 20 Genres in IMDb Top Movies', fontsize=16)
plt.xlabel('Number of Movies', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
genre_counts.head(20)
```

```python
# Calculate average rating
genre_ratings = df2.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False)
# Round it to one decimal place.
genre_d = genre_ratings.round(1)
top_genre_ratings = genre_d.head(20)
top_genre_ratings
```

### Are there correlation between runtime and ratings

```python
df2['Runtime'] = df2['Runtime'].str.replace('min', '', regex=False)
# Ensure the Year column is numeric
df2['Runtime'] = pd.to_numeric(df2['Runtime'], errors='coerce')

```

```python


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df2, x='Runtime', y='IMDB_Rating', alpha=0.6, color='blue')

# Add titles and labels
plt.title('Runtime vs IMDb Ratings', fontsize=16)
plt.xlabel('Runtime (minutes)', fontsize=12)
plt.ylabel('IMDb Rating', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
#Check correlation 
correlation = df2['Runtime'].corr(df2['IMDB_Rating'])
print(f'Correlation between Runtime and Rating: {correlation:.2f}')
```

### Top 10 Directors with Highest Average IMDb Ratings

```python
# Calculate average rating and count of movies for each director
director_stats = df2.groupby('Director')['IMDB_Rating'].agg(['mean', 'count']).reset_index()

director_stats = director_stats.round(1)

# Rename columns for clarity
director_stats.columns = ['Director', 'Average Rating', 'Movie Count']

top_directors = director_stats[director_stats['Movie Count'] > 3]

top_directors_sorted = top_directors.sort_values('Average Rating', ascending=False).head(10)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(data=top_directors_sorted, x='Average Rating', y='Director', palette='crest')

# Add titles and labels
plt.title('Top 10 Directors with Highest Average IMDb Ratings', fontsize=16)
plt.xlabel('Average IMDb Rating', fontsize=12)
plt.ylabel('Director', fontsize=12)

plt.tight_layout()
plt.show()
```
